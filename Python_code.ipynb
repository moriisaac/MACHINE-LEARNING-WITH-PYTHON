````
from sklearn.cluster import KMeans
````

````
from sklearn.metrics import silhouette_score
````

````
import pandas
````

````
import matplotlib.pyplot as plt
````

````
cc_data = pandas.read_csv('cc_data.csv')
````

````
cc_data.dropna(inplace=True)
````

````
dataTrain = cc_data.sample(frac=0.8,random_state=42)
````

````
dataTest = cc_data.drop(dataTrain.index)
````

````
featTraining = dataTrain[['BALANCE', 'BALANCE_FREQUENCY', 'PURCHASES', 'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES',
                       'CASH_ADVANCE', 'PURCHASES_FREQUENCY', 'ONEOFF_PURCHASES_FREQUENCY',
                       'PURCHASES_INSTALLMENTS_FREQUENCY', 'CASH_ADVANCE_FREQUENCY', 'CASH_ADVANCE_TRX',
                       'PURCHASES_TRX', 'CREDIT_LIMIT', 'PAYMENTS', 'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT', 'TENURE']]
````

````
featTraining.dropna(inplace=True)
````

````
NoK = 3
````

````
kmeans_model = KMeans(n_clusters=NoK)
kmeans_model.fit(featTraining)
````

````
LabelsTrain = kmeans_model.predict(featTraining)
````

````
trainSilhoutteScore = silhouette_score(featTraining,LabelsTrain)
````

````
print("Evaluation Based on Silhouette Score:", trainSilhoutteScore)
````

````
featTesting = dataTest[['BALANCE', 'BALANCE_FREQUENCY', 'PURCHASES', 'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES',
                           'CASH_ADVANCE', 'PURCHASES_FREQUENCY', 'ONEOFF_PURCHASES_FREQUENCY',
                           'PURCHASES_INSTALLMENTS_FREQUENCY', 'CASH_ADVANCE_FREQUENCY', 'CASH_ADVANCE_TRX',
                           'PURCHASES_TRX', 'CREDIT_LIMIT', 'PAYMENTS', 'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT', 'TENURE']]
````

````
featTesting.dropna(inplace=True)
````

````
labelsTest = kmeans_model.predict(featTesting)
````

````
silhouetteTest = silhouette_score(featTesting,labelsTest)
````

````
print("Evaluation on Testing, Silhouette Score = ",silhouetteTest)
````

This code defines two functions: `factorial()` and `fibonacci()`. The `factorial()` function calculates the factorial of a number, and the `fibonacci()` function calculates the Fibonacci number for a given number.

The `factorial()` function is recursive, which means that it calls itself to calculate the factorial of a number. The base case is when the number is 0, in which case the function simply returns 1. Otherwise, the function returns the number multiplied by the factorial of the number minus 1.

The `fibonacci()` function is also recursive. The base cases are when the number is 0 or 1, in which case the function simply returns 0 or 1, respectively. Otherwise, the function returns the Fibonacci number for the number minus 1 plus the Fibonacci number for the number minus 2.

The code then prints the factorial of 5 and the Fibonacci number for 10. The factorial of 5 is 120, and the Fibonacci number for 10 is 55.
